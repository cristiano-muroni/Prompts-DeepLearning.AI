 grande qual é o comprimento da saída que você deseja.

 Ok, eu conto três frases, parece que fiz um bom trabalho. E eu também já vi pessoas
 fazerem coisas como, não sei, use no máximo 280 caracteres. Os modelos de linguagem grandes,
 por causa da maneira como interpretam o texto, usando algo chamado tokenização, que não vou falar

````python

import openai

openai.api_key = "HERE YOUR API KEY"

def get_completion(prompt, model = "gpt-3.5-turbo"):
    messages = [ {"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
      model = model,
      messages = messages,
      temperature = 0,
    )
    return response.choices[0].message["content"]
````

````python
fact_sheet_chair = f"""
VISÃO GERAL

- Parte de uma bela linha de móveis para escritório inspirada no design dos anos 50,
incluindo armários de escritório, mesas, estantes e mesas de reunião.
- Várias opções de cores para o revestimento e acabamento da base.
- Disponível com estofamento frontal e traseiro em plástico (SWC-100)
ou estofamento completo (SWC-110) em 10 opções de tecido e 6 opções de couro.
- Opções de acabamento da base: aço inoxidável, preto fosco, branco brilhante ou cromado.
- A cadeira está disponível com ou sem apoios de braços.
- Adequada para ambientes residenciais ou comerciais.
- Certificada para uso em contratos.

CONSTRUÇÃO

- Base de alumínio revestida em plástico com 5 rodízios.
- Cadeira com ajuste pneumático para fácil elevação/descida.

DIMENSÕES

- LARGURA 53 CM | 20.87"
- PROFUNDIDADE 51 CM | 20.08"
- ALTURA 80 CM | 31.32"
- ALTURA DO ASSENTO 44 CM | 16.14"
- PROFUNDIDADE DO ASSENTO 41 CM | 16.14"

OPÇÕES

- Opções de rodízios para piso duro ou macio.
- Duas densidades de espuma do assento disponíveis:
  média (1,8 lb/ft3) ou alta (2,8 lb/ft3).
- Apoios de braços removíveis ou com 8 posições.

MATERIAIS
CONCHA BASE DESLIZANTE

- Alumínio fundido com revestimento de nylon modificado PA6/PA66.
- BASE DESLIZANTE

ASSENTO
- Espuma HD36
PAÍS DE ORIGEM

- Itália.
"""

prompt = f"""
Sua tarefa é ajudar a equipe de marketing a criar uma descrição
para um site de varejo de um produto com base em uma ficha técnica.

Escreva uma descrição do produto com base nas informações fornecidas nas 
especificações técnicas delimitadas por três crases.

A descrição é destinada a revendedores de móveis, portanto, deve ser técnica
e focar nos materiais utilizados na construção do produto.

IMPORTANTE
- Utilize no máximo 90 palavras.
- Exclua da resposta a palavra "traseiro" no texto substituindo por outra palavra
- No começo da descrição inclua um titulo
- No final da descrição inclua ID do produto

Depois da descrição, inclua uma tabela contendo as descrições do produto. A tabela deverá ter duas colunas.
Na primeira coluna inclua o nome da dimensão.
Na segunda coluna inclua as dimensões somente em centímetros

Dê a tabela o título: 'Dimensões do Produto'.
Formate tudo com HTML que possa ser usado em um website.
Coloque dentro da <div> element.


Especificações técnicas: ```{fact_sheet_chair}```.
"""

response = get_completion(prompt)
print(response)

len(response)



````
